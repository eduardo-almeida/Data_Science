```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
temperatura_df = pd.read_csv('data/Celsius-to-Fahrenheit.csv')
```

```python
temperatura_df.head(10)
```

```python
temperatura_df.reset_index(drop= True, inplace= True)
```

```python
temperatura_df.head(10)
```

```python
temperatura_df.info()
```

```python
temperatura_df.describe()
```

```python
sns.scatterplot(temperatura_df['Celsius'], temperatura_df['Fahrenheit']);
```

```python
x_train = temperatura_df['Celsius']
y_train = temperatura_df['Fahrenheit']
```

```python
x_train
```

```python
y_train
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units = 1, input_shape = [1])) 
model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(0.5), loss = 'mean_squared_error')
```

```python
epochs_hist = model.fit(x_train, y_train, epochs = 100)
```

```python
plt.plot(epochs_hist.history['loss']);
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(0.1), loss = 'mean_squared_error')
```

```python
epochs_hist = model.fit(x_train, y_train, epochs = 500)
```

```python
epochs_hist.history.keys()
```

```python
plt.plot(epochs_hist.history['loss'])
plt.title('Modelo de Progresso de Treinamento')
plt.xlabel('pocas')
plt.ylabel('Valor do Erro')

plt.legend(['Valor do Erro']);
```
