## Frequencia dos produtos

```python
df_freq_produto = df.groupby(['descProduto'])[['idTransacao']].count()
```

```python
df_freq_produto
```

```python
df_freq_produto['freq_abs_acum'] = df_freq_produto['idTransacao'].cumsum()
```

```python
df_freq_produto
```

```python
df_freq_produto['freq_relativa'] = df_freq_produto['idTransacao']/df_freq_produto['idTransacao'].sum()
```

```python
df_freq_produto
```

```python
df_freq_produto['freq_relativa_acum'] = df_freq_produto['freq_relativa'].cumsum()
```

```python
df_freq_produto
```
