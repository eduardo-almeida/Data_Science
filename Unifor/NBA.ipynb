### Pipeline para atributos numéricos e categóricos

```python
from sklearn.base import BaseEstimator, TransformerMixin

class CombinedAttributesAdder(BaseEstimator,TransformerMixin):
    def __init__(self):
        return None

    def fit(self, X, y = None):
        return self # Nothing else to do

    def transform(self, X, y = None):

        return self

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipeline = Pipeline([('imputer', SimpleImputer(strategy="mean")),
                         ('scaler', StandardScaler())])

cat_pipeline = Pipeline([('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
                         ('encoder', OneHotEncoder(sparse=False))])

full_pipeline = ColumnTransformer(transformers = [('num_pipeline', num_pipeline, col_num),
                                                  ('cat_pipeline', cat_pipeline, col_cat)])

nba_prepared = full_pipeline.fit_transform(dfLimpo)
print(nba_prepared.shape)
nba_prepared
```

```python
nba_salario  = dfLimpo['salary'].copy()
```

```python
nba_salario.head()
```

```python
from sklearn import metrics
```

## Grid Search

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

rand_forest_reg = RandomForestRegressor()

param_grid = [{'n_estimators': [10, 30, 50], 'max_features': [2, 4, 8, 16, 32, 64]}]  # 3x6=18

grid_search = GridSearchCV(rand_forest_reg,
                           param_grid,
                           scoring='neg_mean_squared_error',
                           refit=True,
                           cv =4, 
                           return_train_score=True) 

grid_search.fit(nba_prepared, nba_salario) #18*4=72 rounds of training!

print(grid_search.best_params_)
```

```python
salarioEsperado =  grid_search.predict(nba_prepared)
```

```python
salarioEsperado
```

```python
print('MAE:', metrics.mean_absolute_error(nba_salario, salarioEsperado).round(8)  )
print('MSE:', metrics.mean_squared_error(nba_salario, salarioEsperado).round(8) )  
print('RMSE:', np.sqrt(metrics.mean_squared_error(nba_salario, salarioEsperado)).round(8))
```

## Randomized Search

```python
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV

rand_forest_reg = RandomForestRegressor()

randomized_search = RandomizedSearchCV(rand_forest_reg,
                                       param_grid,
                                       n_iter=10,
                                       scoring='neg_mean_squared_error',
                                       refit=True,
                                       cv=4, 
                                       random_state=42, 
                                       return_train_score=True) 

randomized_search.fit(nba_prepared, nba_salario) 

print(randomized_search.best_params_)
```

```python
salarioEsperado =  randomized_search.predict(nba_prepared)
```

```python
salarioEsperado
```

```python
print('MAE:', metrics.mean_absolute_error(nba_salario, salarioEsperado).round(8)  )
print('MSE:', metrics.mean_squared_error(nba_salario, salarioEsperado).round(8) )  
print('RMSE:', np.sqrt(metrics.mean_squared_error(nba_salario, salarioEsperado)).round(8))
```
