```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import tensorflow as tf
```

```python
vendas_df = pd.read_csv('data/SalesData.csv')
```

```python
vendas_df.head()
```

```python
vendas_df.info()
```

```python
vendas_df.describe()
```

```python
sns.scatterplot(vendas_df['Temperature'], vendas_df['Revenue']);
```

```python
X_train = vendas_df['Temperature']
Y_train = vendas_df['Revenue']
```

```python
X_train.shape
```

```python
Y_train.head()
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units = 10, input_shape = [1]))
model.add(tf.keras.layers.Dense(units = 1))
```

```python
model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(0.1), loss = 'mean_squared_error')
```

```python
epochs_hist = model.fit(X_train, Y_train, epochs = 1000)
```

```python
epochs_hist.history.keys()
```

```python
plt.plot(epochs_hist.history['loss'])
plt.title('Modelo de Progresso de Treinamento')
plt.xlabel('pocas')
plt.ylabel('Valor do Erro')

plt.legend(['Valor do Erro']);
```

```python
model.get_weights()
```

```python
temp = 5
venda = model.predict([temp])
print('predio de venda = ', venda)
```

```python
plt.scatter(X_train, Y_train, color = 'gray')
plt.plot(X_train, model.predict(X_train), color = 'red')
plt.title('Modelo de Vendas vs Temperatura')
plt.xlabel('Temperatura')
plt.ylabel('Venda');
```
