5- Analisar a necessidade de normalizar as features numéricas (tip: pandas ou scikit-learn)


```python
# Carrega o dataset Ticanic
traindf = pd.read_csv('train.csv')
traindf.info()
```

```python
# Completa os dados faltantes

traindf.drop(['Cabin','Name','PassengerId','Ticket'],axis=1,inplace=True)
avr_age = traindf.Age.mean()
traindf['Age'].fillna(avr_age,inplace=True)
traindf['Embarked'].fillna('S',inplace=True)

traindf.info()
```

```python
#Trnasforma as features categóricas em numéricas

cats = ['Sex','SibSp','Parch','Embarked','Pclass']
train = pd.get_dummies(traindf,columns=cats,drop_first=True).drop(['Survived'],axis=1)
```

```python
train[['Age','Fare']].describe()
```

Analisando os valores max/min das features numéricas acima, vê-se a necessidade de normalização dos dados.

```python
#Normaliza as features numéricas
train.Fare = (train.Fare - train.Fare.mean())/train.Fare.std()
train.Age = (train.Age - train.Age.mean())/train.Age.std()
y_train = traindf['Survived']
```

7- Separar o dataset em treino (75%) / teste (25%) / validação (10% do treino)

```python
msk = np.random.rand(len(train))<=0.75
x = train[msk].values
y = y_train[msk].values
xtest = train[~msk].values
ytest = y_train[~msk].values

msk2 = np.random.rand(len(x))<=0.9
xtrain = x[msk2]
ytrain = y[msk2] 

xval = x[~msk2]
yval = y[~msk2]

```

8- Execute o classificador para 30 k's pulando de 4 em 4 e apresente todas as acurácias utilizando o dataset de validação (Qual o melhor k?) [plotar um gráfico com os resultados]

```python
# Avalia o modelo para vários valores de k

ks = np.arange(1,30)*4-3
scores = []
for k in ks:
    knn = KNNClassifier(k)
    knn.fit(xtrain,ytrain)
    scores.append(np.mean(knn.predict(xval)==yval))
```

```python
# Faz um gráfico das acurácias obtidas no dataset de validação para diferentes valores de k
plt.plot(ks,scores,'r-')
plt.scatter(ks,scores)
plt.xlabel('k')
plt.ylabel('Acurácia')
opt_k = ks[np.argmax(scores)]
print('De acordo com os dados de validação, o k ótimo é',opt_k)
```


8- Executar o classificador para o melhor k encontrado utilizando o dataset de teste e apresentar um relatório da precisão(tip: scikit-learn) [plotar um gráfico com os resultados

```python
from sklearn.metrics import classification_report
knn = KNNClassifier(opt_k)
knn.fit(xtrain,ytrain)
pred = knn.predict(xtest)
print('Report - KNN from scratch')
print(classification_report(ytest,pred))
```

```python
knn_skl = KNeighborsClassifier(n_neighbors = opt_k, weights = 'distance')
knn_skl.fit(xtrain,ytrain)
pred = knn_skl.predict(xtest)
print('Report - KNN sklearn')
print(classification_report(ytest,pred))
```
