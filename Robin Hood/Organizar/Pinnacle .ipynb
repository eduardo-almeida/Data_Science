```python

div_mae = driver.find_element_by_xpath('/html/body/div[2]/div/div[2]/main/div')
html_conteudo = div_mae.get_attribute('outerHTML')
soup = BeautifulSoup(html_conteudo, 'html.parser')
teste = soup.prettify()
```

```python
times = soup.find_all('span', class_='event-row-participant')
valores = soup.find_all('span', class_='style_price__15SlF')
datas = soup.find_all('span', class_='sgl-ParticipantOddsOnly80_Odds')
```

```python
tabela = pd.DataFrame(columns = ['Jogo', 'Vitoria', 'Empate', 'Derrota', 'Local'])
```

```python
for i in range(len(times)):
    if i % 2 == 0:
        listaJogos = times[i].text + " x " + times[i+1].text
        listaVitoria = (valores[int(7*i/2)].text)
        listaEmpates = (valores[int(7*i/2+1)].text)
        listaDerrota = (valores[int(7*i/2+2)].text)
        tabela = tabela.append({'Jogo' : listaJogos,
                                'Vitoria' : listaVitoria,
                                'Empate' : listaEmpates,
                                'Derrota' : listaDerrota,
                                'Local' : 'Pinnacle'}, 
                        ignore_index=True)
```
