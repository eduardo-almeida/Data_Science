# Posições do basquete:

### 1. **Shooting Guard (SG) - Ala-armador**
- **Função principal**: Pontuar, especialmente com arremessos de longa distância.
- **Características**: Bom arremessador, ágil, capaz de criar suas próprias jogadas e jogar sem a bola.
- **Exemplo**: Michael Jordan, Kobe Bryant.

### 2. **Point Guard (PG) - Armador**
- **Função principal**: Liderar a equipe em quadra, distribuindo a bola e organizando as jogadas.
- **Características**: Boa visão de jogo, habilidade de passe, controle de bola e capacidade de tomar decisões rápidas.
- **Exemplo**: Magic Johnson, Stephen Curry.

### 3. **Center (C) - Pivô**
- **Função principal**: Jogar próximo à cesta, tanto na defesa quanto no ataque. Defende o garrafão e pega rebotes.
- **Características**: Alto, forte, bom reboteiro e defensor de aro. Capaz de pontuar perto da cesta.
- **Exemplo**: Shaquille O'Neal, Kareem Abdul-Jabbar.

### 4. **Power Forward (PF) - Ala-pivô**
- **Função principal**: Jogar próximo à cesta, ajudando nos rebotes e na defesa, mas também capaz de fazer arremessos de média distância.
- **Características**: Forte, bom reboteiro, pode jogar tanto dentro quanto fora do garrafão.
- **Exemplo**: Tim Duncan, Karl Malone.

### 5. **Small Forward (SF) - Ala**
- **Função principal**: Versátil, pode pontuar de diversas maneiras e defender múltiplas posições.
- **Características**: Atlético, bom arremessador, capaz de infiltrar e defender bem.
- **Exemplo**: LeBron James, Larry Bird.

### 6. **Guard (G) - Armador/Ala-armador**
- **Função principal**: Pode desempenhar tanto a função de armador quanto de ala-armador, dependendo da necessidade da equipe.
- **Características**: Versátil, bom controle de bola e arremesso, visão de jogo.
- **Exemplo**: Jogadores que podem alternar entre PG e SG.

### 7. **Forward (F) - Ala/Ala-pivô**
- **Função principal**: Pode desempenhar tanto a função de ala quanto de ala-pivô, dependendo da necessidade da equipe.
- **Características**: Versátil, bom na defesa e no ataque, capaz de jogar dentro e fora do garrafão.
- **Exemplo**: Jogadores que podem alternar entre SF e PF.

### 8. **Guard-Forward (GF) - Armador-Ala**
- **Função principal**: Pode desempenhar funções de armador, ala-armador e ala, dependendo da necessidade da equipe.
- **Características**: Muito versátil, capaz de defender múltiplas posições e contribuir de várias formas no ataque.
- **Exemplo**: Jogadores que são bons tanto no perímetro quanto nas infiltrações e defesa.

As médias salariais dos jogadores da NBA podem variar amplamente, dependendo de fatores como o talento individual, a experiência e a demanda de mercado.
No entanto, geralmente, a ordem das posições de acordo com a média salarial, da maior para a menor, tende a ser a seguinte:

1. **Point Guard (PG) - Armador**
2. **Shooting Guard (SG) - Ala-armador**
3. **Small Forward (SF) - Ala**
4. **Power Forward (PF) - Ala-pivô**
5. **Center (C) - Pivô**

### Explicação

- **Point Guards (PGs)** e **Shooting Guards (SGs)** frequentemente têm salários altos porque são responsáveis pela criação de jogadas e pontuação, posições que demandam muita habilidade e responsabilidade.
- **Small Forwards (SFs)** também tendem a ter salários elevados devido à sua versatilidade e capacidade de impactar o jogo tanto ofensivamente quanto defensivamente.
- **Power Forwards (PFs)**, embora essenciais, geralmente têm salários um pouco mais baixos que os PGs e SGs, mas ainda são bastante valorizados.
- **Centers (Cs)**, tradicionalmente, tinham altos salários, mas com a evolução do jogo para um estilo mais ágil e de perímetro, os valores têm variado mais amplamente.

Posições como **Guard (G)**, **Forward (F)** e **Guard-Forward (GF)** são híbridas e podem variar bastante em termos de salário, dependendo de como são utilizados pelas equipes e de suas habilidades específicas.

Essa ordem é baseada em tendências gerais e pode variar ano a ano, dependendo do desempenho de jogadores específicos e das necessidades das equipes.


### Estatísticas Gerais

1. **Games Played (Jogos Jogados)**
   - **Definição**: Número total de jogos em que o jogador participou.
   - **Importância**: Indica a disponibilidade e durabilidade do jogador ao longo da temporada.

2. **Avg Minutes (Média de Minutos)**
   - **Definição**: Média de minutos jogados por jogo.
   - **Importância**: Mostra quanto tempo o jogador está em quadra, refletindo sua importância para a equipe.

3. **Minutes (Minutos)**
   - **Definição**: Total de minutos jogados ao longo da temporada.
   - **Importância**: Indica o tempo total de participação em jogos.

### Estatísticas de Faltas

4. **Avg Fouls (Média de Faltas)**
   - **Definição**: Média de faltas cometidas por jogo.
   - **Importância**: Ajuda a avaliar a disciplina do jogador e sua tendência a cometer faltas.

5. **Flagrant Fouls (Faltas Flagrantes)**
   - **Definição**: Faltas graves cometidas durante a temporada.
   - **Importância**: Faltas flagrantes são faltas violentas ou desnecessárias, indicando comportamento imprudente.

6. **Technical Fouls (Faltas Técnicas)**
   - **Definição**: Faltas atribuídas por comportamento antidesportivo ou infrações técnicas.
   - **Importância**: Refletem problemas de conduta ou desobediência às regras.

7. **Ejections (Expulsões)**
   - **Definição**: Número de vezes que o jogador foi expulso de um jogo.
   - **Importância**: Indica infrações graves ou repetidas durante os jogos.

### Estatísticas de Desempenho

8. **Double-Double**
   - **Definição**: Jogos em que o jogador registrou dois dígitos em duas das cinco principais categorias estatísticas (pontos, rebotes, assistências, roubos de bola, tocos).
   - **Importância**: Mostra versatilidade e consistência em várias áreas do jogo.

9. **Triple-Double**
   - **Definição**: Jogos em que o jogador registrou dois dígitos em três das cinco principais categorias estatísticas.
   - **Importância**: Reflete um desempenho excepcional em múltiplas áreas.

### Estatísticas de Rebote

10. **Rebounds (Rebotes)**
    - **Definição**: Total de rebotes (ofensivos e defensivos) pegos pelo jogador.
    - **Importância**: Indica a habilidade do jogador em recuperar a bola após um arremesso perdido.

11. **Avg Rebounds (Média de Rebotes)**
    - **Definição**: Média de rebotes por jogo.
    - **Importância**: Mostra a consistência do jogador em pegar rebotes.

### Estatísticas de Pontuação

12. **Avg Points (Média de Pontos)**
    - **Definição**: Média de pontos marcados por jogo.
    - **Importância**: Mostra a eficiência e regularidade do jogador em pontuar.

### Estatísticas de Arremesso

13. **Avg Field Goals Made (Média de Cestas de Campo Feitas)**
    - **Definição**: Média de cestas de campo convertidas por jogo.
    - **Importância**: Reflete a habilidade do jogador em converter arremessos.

14. **Avg Field Goals Attempted (Média de Cestas de Campo Tentadas)**
    - **Definição**: Média de arremessos de campo tentados por jogo.
    - **Importância**: Mostra a frequência com que o jogador arremessa.

15. **Field Goal Pct (Percentual de Cestas de Campo)**
    - **Definição**: Percentual de arremessos de campo convertidos em relação ao total de tentativas.
    - **Importância**: Indica a eficiência do jogador em arremessar.

### Estatísticas de Três Pontos

16. **Avg Three Point Field Goals Made (Média de Cestas de Três Pontos Feitas)**
    - **Definição**: Média de cestas de três pontos convertidas por jogo.
    - **Importância**: Reflete a habilidade do jogador em converter arremessos de longa distância.

17. **Avg Three Point Field Goals Attempted (Média de Cestas de Três Pontos Tentadas)**
    - **Definição**: Média de arremessos de três pontos tentados por jogo.
    - **Importância**: Mostra a frequência com que o jogador tenta arremessos de três pontos.

18. **Three Point Field Goal Pct (Percentual de Cestas de Três Pontos)**
    - **Definição**: Percentual de arremessos de três pontos convertidos em relação ao total de tentativas.
    - **Importância**: Indica a eficiência do jogador em arremessos de longa distância.

### Estatísticas de Lance Livre

19. **Avg Free Throws Made (Média de Lances Livres Feitos)**
    - **Definição**: Média de lances livres convertidos por jogo.
    - **Importância**: Reflete a habilidade do jogador em converter lances livres.

20. **Avg Free Throws Attempted (Média de Lances Livres Tentados)**
    - **Definição**: Média de lances livres tentados por jogo.
    - **Importância**: Mostra a frequência com que o jogador tenta lances livres.

21. **Free Throw Pct (Percentual de Lances Livres)**
    - **Definição**: Percentual de lances livres convertidos em relação ao total de tentativas.
    - **Importância**: Indica a eficiência do jogador na linha de lance livre.

22. **Free Throws Made (Lances Livres Feitos)**
    - **Definição**: Total de lances livres convertidos.
    - **Importância**: Mostra o total de lances livres bem-sucedidos.

23. **Free Throws Attempted (Lances Livres Tentados)**
    - **Definição**: Total de lances livres tentados.
    - **Importância**: Indica a agressividade do jogador em tentar lances livres.

### Estatísticas de Assistências e Erros

24. **Avg Assists (Média de Assistências)**
    - **Definição**: Média de assistências por jogo.
    - **Importância**: Mostra a consistência do jogador em criar oportunidades de pontuação para os companheiros.

25. **Avg Turnovers (Média de Erros)**
    - **Definição**: Média de erros (perdas de posse de bola) por jogo.
    - **Importância**: Indica a frequência com que o jogador comete erros.

26. **Assists (Assistências)**
    - **Definição**: Total de assistências feitas pelo jogador durante a temporada.
    - **Importância**: Reflete a capacidade do jogador de distribuir a bola para seus companheiros de equipe.

27. **Turnovers (Erros)**
    - **Definição**: Total de erros (perdas de posse de bola) cometidos pelo jogador durante a temporada.
    - **Importância**: Indica a propensão do jogador a cometer erros durante o jogo.

### Estatísticas de Pontuação

28. **Points (Pontos)**
    - **Definição**: Total de pontos marcados pelo jogador durante a temporada.
    - **Importância**: Principal indicador da capacidade ofensiva do jogador.

29. **Field Goals Made (Cestas de Campo Feitas)**
    - **Definição**: Total de cestas de campo convertidas pelo jogador durante a temporada.
    - **Importância**: Indica a eficácia do jogador em converter arremessos de campo.

30. **Field Goals Attempted (Cestas de Campo Tentadas)**
    - **Definição**: Total de arremessos de campo tentados pelo jogador durante a temporada.
    - **Importância**: Mostra a agressividade do jogador em tentar arremessos.

31. **Three Point Field Goals Made (Cestas de Três Pontos Feitas)**
    - **Definição**: Total de cestas de três pontos convertidas pelo jogador durante a temporada.
    - **Importância**: Reflete a habilidade do jogador em converter arremessos de longa distância.

32. **Three Point Field Goals Attempted (Cestas de Três Pontos Tentadas)**
    - **Definição**: Total de arremessos de três pontos tentados pelo jogador durante a temporada.
    - **Importância**: Mostra a frequência com que o jogador tenta arremessos de três pontos.

33. **Total Points (Pontos Totais)**
    - **Definição**: Total acumulado de pontos marcados pelo jogador durante a temporada.
    - **Importância**: Indica a contribuição total do jogador na pontuação da equipe.

### Estatísticas Defensivas

34. **Avg Steals (Média de Roubos de Bola)**
    - **Definição**: Média de roubos de bola por jogo.
    - **Importância**: Mostra a consistência do jogador em roubar a bola do time adversário.

35. **Avg Blocks (Média de Tocos)**
    - **Definição**: Média de tocos por jogo.
    - **Importância**: Reflete a habilidade do jogador em bloquear os arremessos adversários.

36. **Steals (Roubos de Bola)**
    - **Definição**: Total de roubos de bola pelo jogador durante a temporada.
    - **Importância**: Indica a capacidade do jogador de interceptar passes e criar transições rápidas.

37. **Blocks (Tocos)**
    - **Definição**: Total de tocos pelo jogador durante a temporada.
    - **Importância**: Mostra a habilidade do jogador em defender a cesta bloqueando os arremessos adversários.
