```python
import pandas as pd
import pandasql as ps
```

```python
pedido = pd.read_csv("../data/pedido.csv")
item = pd.read_csv("../data/item.csv")
```

```python
pedido.head()
```

```python
item.head()
```

```python
tb_activate = ps.sqldf("""
                        select 
                            t2.idVendedor,
                            min(date(dtPedido)) as dtAtivacao
                        
                        from pedido as t1
                        
                        left join item as t2
                        on t1.idpedido = t2.idPedido                        
                        
                        where t1.dtPedido >= '2018-01-01'
                        and t1.dtPedido <= '2018-02-16'
                        and t2.idVendedor is not null
                        
                        group by t2.idVendedor
                        order by dtAtivacao
""")
```

```python
tb_teste = ps.sqldf("""
                        select 
                            t1.*,
                            t2.*,
                            t3.*,
                            t4.*,
                            t5.*,
                            t6.*,
                            case when t7.idVendedor is null then 1 else 0 end as flChurn,
                            t7.dtAtivacao
                        
                        from vendedorVenda as t1
                        
                        left join vendedorAvaliacao as t2
                        on t1.idVendedor = t2.idVendedor
                        and t1.dtReference = t2.dtReference
                        
                        left join vendedorCliente as t3
                        on t1.idVendedor = t3.idVendedor
                        and t1.dtReference = t3.dtReference
                        
                        left join vendedorEntrega as t4
                        on t1.idVendedor = t4.idVendedor
                        and t1.dtReference = t4.dtReference
                        
                        left join vendedorPagamento as t5
                        on t1.idVendedor = t5.idVendedor
                        and t1.dtReference = t5.dtReference
                        
                        left join vendedorProduto as t6
                        on t1.idVendedor = t6.idVendedor
                        and t1.dtReference = t6.dtReference
                        
                        left join tb_activate as t7
                        on t1.idVendedor = t7.idVendedor
                        and (date(t7.dtAtivacao) - date(t1.dtReference) + t1.qtdRecencia) <= 45
                        
                        where t1.qtdRecencia <= 45
                        
                        
""")
```
