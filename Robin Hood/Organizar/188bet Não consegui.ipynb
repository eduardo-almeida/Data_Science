```python
driver = webdriver.Firefox()
driver.get("https://www.188bet.com/pt-br/sports/match-by-sport/football?coupon=11")
```

```python
div_mae = driver.find_element_by_xpath('/html/body/div[3]/div/div[1]/div[2]/div/div[1]/div[2]/div/div[1]/div[2]/div[2]')
html_conteudo = div_mae.get_attribute('outerHTML')
soup = BeautifulSoup(html_conteudo, 'html.parser')
teste = soup.prettify()
```

```python
print(teste)
```

```python
times = soup.find_all('span', class_='teamNameFirstPart')
valores = soup.find_all('div', class_='odds')
datas = soup.find_all('span', class_='sgl-ParticipantOddsOnly80_Odds')
```

```python
weather = pd.DataFrame({
        "casa": valores
    })
```

```python
tabela = pd.DataFrame(columns = ['Jogo', 'Vitoria', 'Empate', 'Derrota', 'Local'])
```

```python
for i in range(len(times)):
    if i % 2 == 0:
        listaJogos = times[i].text , " x " , times[i+1].text
        listaVitoria = (valores[int(i/2)].text)
        listaEmpates = (valores[int(i/2 + len(times)/2)].text)
        listaDerrota = (valores[int(i/2 + 2*len(times)/2)].text)
        tabela = tabela.append({'Jogo' : listaJogos,
                                'Vitoria' : listaVitoria,
                                'Empate' : listaEmpates,
                                'Derrota' : listaDerrota,
                                'Local' : 'Pinnacle'}, 
                        ignore_index=True)
```
