```python
tabela = pd.DataFrame(columns = ['Jogo', 'Vitoria', 'Empate', 'Derrota', 'Local'])
```

```python
for i in range(len(times)):
    if i % 2 == 0:
        listaJogos = (times[i].text).replace('\n', '') + " x " + (times[i+1].text).replace('\n', '')
        listaVitoria = (valores[int(5*i/2+2)].text).replace('\n', '')
        listaEmpates = (valores[int(5*i/2+3)].text).replace('\n', '')
        listaDerrota = (valores[int(5*i/2+4)].text).replace('\n', '')
        tabela = tabela.append({'Jogo' : listaJogos,
                                'Vitoria' : listaVitoria,
                                'Empate' : listaEmpates,
                                'Derrota' : listaDerrota,
                                'Local' : 'Betfair'}, 
                        ignore_index=True)
```

```python
tabela["Jogo"] = tabela["Jogo"].replace ("Atlético-MG", "Terst-MG", regex=True)
```
