```python
import requests
from bs4 import BeautifulSoup

# URL da página principal
url = "https://www.arquidiocesedefortaleza.org.br/arquidiocese/regioes/regiao-metropolitana-sao-jose/paroquias-da-regiao-sao-jose/"

# Fazer a requisição HTTP para obter o HTML da página principal
response = requests.get(url)
response.raise_for_status()  # Verificar se a requisição foi bem-sucedida

# Parsear o HTML da página principal
soup = BeautifulSoup(response.text, 'html.parser')

# Encontrar todos os links dentro das tags <a> na página principal
links = [a['href'] for a in soup.find_all('a', href=True) if a['href'].startswith('https')]

# Função para obter endereço e bairro de uma página
def get_address_and_neighborhood(url):
    response = requests.get(url)
    response.raise_for_status()
    page_soup = BeautifulSoup(response.text, 'html.parser')
    
    endereco_label = page_soup.find('td', text='Endereço:')
    bairro_label = page_soup.find('td', text='Bairro:')
    
    endereco = endereco_label.find_next_sibling('td').text if endereco_label else 'Endereço não encontrado'
    bairro = bairro_label.find_next_sibling('td').text if bairro_label else 'Bairro não encontrado'
    
    return endereco, bairro

# Iterar sobre os links e pegar o endereço e bairro
for link in links:
    endereco, bairro = get_address_and_neighborhood(link)
    print(f'Link: {link}')
    print(f'Endereço: {endereco}')
    print(f'Bairro: {bairro}')
    print('-----------------------------')
```
